```python
import numpy as np
np.random.seed(0)
vx = np.random.rand(5,5,5) + 1j*np.random.rand(5,5,5)
vy = np.random.rand(5,5,5) + 1j*np.random.rand(5,5,5)
vz = np.random.rand(5,5,5) + 1j*np.random.rand(5,5,5)
print(vx, vy, vz)
```

```python
fvx = np.fft.fftn(vx)
fvy = np.fft.fftn(vy)
fvz = np.fft.fftn(vz)
fv = np.abs(fvx)**2 + np.abs(fvy)**2 + np.abs(fvz)**2
```
