```python
pos = np.array([[1/3,1/3,1/3]])
mass = np.array([1e-3])
vx = np.array([1])

Nsize = 3
# initialize the catalog
cat = ArrayCatalog({'Position':pos, 'vx':vx, 'Masses':mass})
cat.attrs['BoxSize'] = 1.
mesh_vx = cat.to_mesh(value='vx', Nmesh=Nsize,
    resampler='tsc', compensated=False, interlaced=False)
grid_vx = mesh_vx.paint()

# Plot
plt.imshow(grid_vx[:,:,1])
plt.colorbar()
plt.show()
print(grid_vx)
```
