What happens at the edge of the sphere when we use the mv/m schematics?

```python
pos = np.array([[1.4,0.5,0.5],[0.6,0.5,0.5]])
R = np.array([0.4,0.4])
rho = np.array([1,1])
vx = np.array([0.5, 0.7])

# Voxelize
res = 64
grid_rhovx = Voxelize.__call__(self=Voxelize, box_L=1, coords=pos, 
                               radii=R, field=rho*vx, box=res)
grid_rho = Voxelize.__call__(self=Voxelize, box_L=1, coords=pos, 
                               radii=R, field=rho, box=res)
zslice = int(0.5 * res)
plt.imshow((grid_rhovx/grid_rho)[:,:,zslice])
plt.colorbar()
```

$$ \sum_\alpha \frac{f_i V_{i,\alpha}}{V_\alpha} $$

```python
# 1 particle in 1 cell
pos = np.array([[0.2,0.5,0.5],[0.8,0.5,0.5]])
R = np.array([0.1,0.1])
V = 4/3 * np.pi * R**3

rho = np.array([1.,1.])
rho_V = rho / V
grid = Voxelize.__call__(self=Voxelize, box_L=1, coords=pos, radii=R, field=rho_V, box=3)

Lcell = 1./3
grid *= Lcell**3
print(grid)
plt.imshow(grid[:,:,1])
plt.colorbar()
```
